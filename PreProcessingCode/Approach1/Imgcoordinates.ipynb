```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image  
import json


json_face_path = '/Users/chedvi/Downloads/SER 517/GazeCapture/00002/appleFace.json'
json_face_path1 = '/Users/chedvi/Downloads/SER 517/GazeCapture/00002/appleLeftEye.json'
json_face_path2 = '/Users/chedvi/Downloads/SER 517/GazeCapture/00002/appleRightEye.json'
json_face_path3 = '/Users/chedvi/Downloads/SER 517/GazeCapture/00002/faceGrid.json'
image_path = '/Users/chedvi/Downloads/SER 517/GazeCapture/00002/frames/00000.jpg'

# Open the file and load its contents as a dictionary


def draw_bounding_box_on_image(image_path, json_face_path, json_face_path1, json_face_path2 ):

    with open(json_face_path, 'r') as file:
        json_data = json.load(file)

    with open(json_face_path1, 'r') as file:
        json_data1 = json.load(file)

    with open(json_face_path2, 'r') as file:
        json_data2 = json.load(file)
        
    with open(json_face_path3, 'r') as file:
        json_data3 = json.load(file)




    left_top_x = json_data['X'][0]
    left_top_y = json_data['Y'][0]
    box_width = json_data['W'][0]
    box_height = json_data['H'][0]




    left_top_x1 = json_data1['X'][0]
    left_top_y1 = json_data1['Y'][0]
    box_width1 = json_data1['W'][0]
    box_height1 = json_data1['H'][0]


    left_top_x2 = json_data2['X'][0]
    left_top_y2 = json_data2['Y'][0]
    box_width2 = json_data2['W'][0]
    box_height2 = json_data2['H'][0]
    
    
    left_top_x3 = json_data3['X'][0]
    left_top_y3 = json_data3['Y'][0]
    box_width3 = json_data3['W'][0]
    box_height3 = json_data3['H'][0]
    
    
    
    
    
    
    # Load the image
    img = Image.open(image_path)

    # Create figure and axes
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(img)

    # Add the bounding box to the image
    rect = patches.Rectangle((left_top_x, left_top_y), box_width, box_height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    rect = patches.Rectangle((left_top_x + left_top_x1, left_top_y + left_top_y1), box_width1, box_height1, linewidth=1, edgecolor='b', facecolor='none')
    ax.add_patch(rect)
    rect = patches.Rectangle((left_top_x + left_top_x2, left_top_y + left_top_y2), box_width2, box_height2, linewidth=1, edgecolor='g', facecolor='none')
    ax.add_patch(rect)
    
    rect = patches.Rectangle((left_top_x3, left_top_y3), box_width3, box_height3, linewidth=1, edgecolor='y', facecolor='none')
    ax.add_patch(rect)
    
    
    plt.axis('off')  # Turn off axis labels and ticks
    plt.show()


    # Assuming you have the rectangle coordinates (left_top_x, left_top_y), width, and height
    # Let's say the face part of the image is stored in a variable called 'face_image'

    # Get the region of interest (ROI) corresponding to the face
    face_roi = img.crop((left_top_x, left_top_y, left_top_x + box_width, left_top_y + box_height))

    # Resize the ROI to 25x25
    resized_face = face_roi.resize((25, 25))

    
        # Create figure and axes
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(resized_face)
    rect = patches.Rectangle((left_top_x3, left_top_y3), box_width3, box_height3, linewidth=1, edgecolor='y', facecolor='none')
    ax.add_patch(rect)
    





draw_bounding_box_on_image(image_path, json_face_path,json_face_path1,json_face_path2)

# l_eye_det = json.load(open('/appleLeftEye.json'))
# r_eye_det = json.load(open('/appleRightEye.json'))
# facegrid = json.load(open('/appleRightEye.json'))
# dot = json.load(open('/dotInfo.json'))

# portrait_orientation = np.asarray(screen_info["Orientation"])==1
# l_eye_valid, r_eye_valid = np.array(l_eye_det['IsValid']), np.array(r_eye_det['IsValid'])
# valid_ids = l_eye_valid*r_eye_valid*portrait_orientation
```
